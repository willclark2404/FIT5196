```python
titlelist = []
for elem in tree.findall('Term/Title'):
       titlelist.append(elem.text)
len(titlelist)
```

```python
descriptionlist = []

for elem in tree.findall('Term/Title'):
     for i in elem.getparent().iter(tag = 'Description'):
            descriptionlist.append(i.text)
len(descriptionlist)

#Checked that .iter command worked, then found alternative methods for trickier data structures
```

```python
relatedlist = []

for elem in tree.findall('Term/Title'):
    relatedlist.append(elem.getparent().findall('RelatedTerms/Term/Title'))
len(relatedlist)

# for this I had to search by the title, such that children of the same parent node would yield the same number of arrays
```

```python
print(relatedlist[:5])
type(relatedlist[0][0])
#checked that this created the result as follows; 2 or more child elements would be grouped together, and 1 child would get their own list. So far, so good, len matches for all columns
```

```python
relationshiplist = []

for elem in tree.findall('Term/Title'):
    relationshiplist.append(elem.getparent().findall('RelatedTerms/Term/Relationship'))
len(relationshiplist)
```

```python
description = []
for i in descriptionlist:
    description.append(i)
print(len(description))
description[:5]
```

```python
related = []

for i in relatedlist:
    if len(i) == 1:
            i[0] = i[0].text
            related.append(i)
    else:
        related.append(i)
            


len(related)

#now converting elements into their text forms
```

```python
print(related[:10])
```

```python
for i in related:
    pos=0
    if len(i)>1:
        while pos < len(i):
            i[pos] = i[pos].text
            pos+=1

print(related[4][0:3])

#nested elements for groups of 2 or more children required a looped iteration conversion
```

```python
relationship = []

for i in relationshiplist:
    if len(i) == 1:
            i[0] = i[0].text
            relationship.append(i)
    else:
        relationship.append(i)
            


len(relationship)
```

```python
for i in relationship:
    pos=0
    if len(i)>1:
        while pos < len(i):
            i[pos] = i[pos].text
            pos+=1
```

```python
print(len(relationship[5]))
print(relationship[4:8])
```

```python
dataDict = {}
dataDict['Title'] = titlelist
dataDict['Description'] = description
dataDict['Related Terms'] = related
dataDict['Relationship'] = relationship

df = pd.DataFrame(dataDict, index = titlelist)
df.index.name = 'Title'
df.head()
```
